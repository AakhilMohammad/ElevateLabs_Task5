```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv(r"C:\Users\AAKHIL MOHAMMAD\Downloads\cleaned_titanic.csv")

# Create correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Distribution of Numeric Features
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
for col in numeric_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
#  Boxplots by Survival Status
for col in ['Age', 'Fare']:
    plt.figure()
    sns.boxplot(x='Survived', y=col, data=df)
    plt.title(f'{col} by Survival')
    plt.show()
```

```python
# Pairplot of Selected Features
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.suptitle("Pairplot of Survival vs Selected Features", y=1.02)
plt.show()
```

```python
# Categorical Feature Impact on Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Count by Passenger Class")
plt.show()
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Sex")
plt.show()
```

```python
# 🧾 Summary of Findings
print("""
- Survival is strongly related to Passenger Class and Sex
- Fare tends to be higher for survivors
- Younger passengers show slightly better survival rates
- Use these features for further modeling or decision-making
""")
```
